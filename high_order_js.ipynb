```python
%%writefile code.js

// Part A
const data = [5, 12, 3, 20, 11, 9, 30];

const loop = [];
for (let i = 0; i < data.length; i++) {
  if (data[i] > 10) {
    loop.push(data[i] * 2);
  }
}

// Pipeline
const pipe = data
  .filter(value => value > 10)
  .map(value => value * 2);

// Console
console.log("Part A:");
console.log("Loop result:", loop);
console.log("Pipeline result:", pipe);
console.log("Results match:", JSON.stringify(loop) === JSON.stringify(pipe));
console.log();
console.log();



// Part B
const numbers = [1, 2, 3, 4, 5, 6];

// Pipeline
const sumOfDoubledEvens = numbers
  .filter(num => num % 2 === 0)         // filter evens
  .map(num => num * 2)                  // map to doubled
  .reduce((sum, num) => sum + num, 0);  // reduce to sum

// Console
console.log("Part B:");
console.log("Sum of doubled evens:", sumOfDoubledEvens);
console.log("Expected: 24");
console.log();
console.log();



// Part C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

console.log("Part C:");

// Any user under 18
const minor = users.some(user => user.age < 18);
console.log("Is any user under 18?", minor);
console.log();

// Find id === 3
const id3 = users.find(user => user.id === 3);
console.log("User with id === 3:", id3);
console.log();

// Print each user separately
console.log("User names:");
users.forEach(user => console.log(user.name));
console.log();
console.log();



// Part D
function repeat(n, action) {
  for (let i = 0; i < n; i++) {
    action(i);
  }
}

console.log("Part D:");

repeat(3, console.log);
repeat(4, i => console.log("Step " + i));

```
